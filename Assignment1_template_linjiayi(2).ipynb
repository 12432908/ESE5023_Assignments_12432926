```python
import math

# def calculate_F_values(N_list):
#     # Your code here
def F(x):
    if x == 1:
        return 1
    else:
        return F(math.ceil(x / 3)) + 2*x

list_a = [F(i) for i in range(1, 11)]
print(list_a)
```

**Notice:** for a given cell, you are **only allowed** to move either rightward or downward.

```python
# def Count_path(matrix):
    # Your code here
#5.2 从csdn中寻找灵感，问了助教吴星沂以及ai更改错误
def Count_path(arr_1):
    row, column = arr_1.shape#获取输入数组的维度 
    path_count = np.zeros((row, column), dtype = int)#设置全为0的新的矩阵，用来储存路径
    path_count[0][0] = 1#起始路径初始化
    
    for i in range(row):
        for j in range(column):
            if arr_1[i][j] == 1:
                if i > 0 and arr_1[i - 1][j] == 1 :
                    path_count[i][j] += path_count[i - 1][j]
                if j > 0 and arr_1[i][j - 1] == 1 :
                        path_count[i][j] += path_count[i][j - 1]
                
    return path_count



arr_1 = np.random.randint(0, 2,(4, 4))
arr_1[0][0] = 1
arr_1[3][3] = 1
print(arr_1)
print(Count_path(arr_1))
print(Count_path(arr_1)[3][3])



```
